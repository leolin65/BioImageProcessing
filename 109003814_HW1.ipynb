```python
# %load demo_image_display.py
#!/usr/bin/env python

# This is a demo to show how to display an image using python.

# In[1]:


import numpy as np
import matplotlib.pyplot as plt
# import required image module
from PIL import Image

# open the binary data file as binary_file
with open("mri-a.raw", "rb") as binary_file:
    #read in the entire data set as an 1-d array
    data = np.fromfile(binary_file, dtype=np.ubyte)

# reshape the data into row * col 
data = data.reshape([256,256])


#display the image, and use the color map gray
plt.imshow(data, cmap='gray')
plt.show()
```

```python
#1.(b) (Image Inversion) Subtract each pixel from 255 and store the results.
x = np.subtract(256, data)
plt.imshow(x,cmap='gray')
plt.show()
```

```python
#1.(c) Image transpose
x = np.transpose(data)
plt.imshow(x, cmap='gray')
plt.show()
```

```python
#1.(d) Flip image vertically - Method 1

print(data.shape)
# (225, 400, 3)
#np.flipud ==> Reverse the order of elements along axis 0 (up/down).
x1 =Image.fromarray(np.flipud(data))

plt.imshow(x1, cmap='gray')
plt.show()

```

```python
#1(e) Flip image horizontally
x2 = Image.fromarray(np.fliplr(data))
plt.imshow(x2, cmap='gray')
plt.show()


```

```python
#1.(d) Flip image - Method 2
import cv2

img_flip_ud = cv2.flip(data, 0)
plt.imshow(img_flip_ud, cmap='gray')
plt.show()

img_flip_lr = cv2.flip(data, 1)
plt.imshow(img_flip_lr, cmap='gray')
plt.show()

# True

img_flip_ud_lr = cv2.flip(data, -1)
plt.imshow(img_flip_ud_lr, cmap='gray')
plt.show()

# True
```
